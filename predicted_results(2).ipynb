```python
def extract_features(audio_path):
    try:
        y, sr = librosa.load(audio_path, duration=30)

        # Extract MFCCs, Chroma, and Mel Spectrogram features
        mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).T, axis=0)
        chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr).T, axis=0)
        mel = np.mean(librosa.feature.melspectrogram(y=y, sr=sr).T, axis=0)

        return mfccs, chroma, mel

    except FileNotFoundError:
        print(f"File not found: {audio_path}")
        return None, None, None

    except Exception as e:
        print(f"An error occurred while processing {audio_path}: {e}")
        return None, None, None
```

# Prediction Music Attribute Function

```python

def predict_music_attributes(audio_files):
    predictions = []

    for audio_path in audio_files:
        mfccs, chroma, mel = extract_features(audio_path)
        if mfccs is not None and chroma is not None and mel is not None:
            mfccs = np.pad(mfccs, (0, max(0, 100 - len(mfccs))), 'constant')[:100]
            chroma = np.pad(chroma, (0, max(0, 100 - len(chroma))), 'constant')[:100]
            mel = np.pad(mel, (0, max(0, 100 - len(mel))), 'constant')[:100]
            features = np.hstack([mfccs, chroma, mel])
            X_new = np.array(features).reshape((1, 1, -1))

            # Predict for each model
            prediction = {}
            prediction['Track'] = os.path.basename(audio_path)
            for attribute, model in models.items():
                y_pred = model.predict(X_new)
                if attribute in label_encoders:  # Categorical attribute
                    y_pred_classes = np.argmax(y_pred, axis=1)
                    prediction[attribute] = label_encoders[attribute].inverse_transform(y_pred_classes)[0]

                else:  # Numerical attribute
                    prediction[attribute] = y_pred[0][0]

            predictions.append(prediction)

    # Convert the predictions to a DataFrame
    return pd.DataFrame(predictions)

# audio_files = ['/path/to/song1.mp3', '/path/to/song2.mp3']
# results_df = predict_music_attributes(audio_files)
# print(results_df)
```
